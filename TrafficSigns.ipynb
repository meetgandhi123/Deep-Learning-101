```
from keras.models import Sequential
from keras.layers import Dense, Conv2D, Flatten
model = Sequential()
model.add(Conv2D(3000, kernel_size=3,strides=2, activation='relu', input_shape=(50,50,3)))
model.add(Conv2D(2000, kernel_size=3,strides=2, activation='relu'))
model.add(Conv2D(1000, kernel_size=3, activation='relu'))
model.add(Conv2D(500, kernel_size=3, activation='relu'))
model.add(Conv2D(100, kernel_size=3, activation='relu'))
model.add(Flatten())
model.add(Dense(43, activation='softmax'))
model.summary()
```

```
#model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```
from keras.optimizers import Adam,SGD,Adagrad,Adadelta,RMSprop

adam = Adam(epsilon = 0.01)
model.compile(optimizer = adam, loss = 'categorical_crossentropy', metrics = ['accuracy'])

from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale=1./255)

test_datagen = ImageDataGenerator(rescale=1./255)


train_folder = '/content/Train' 
test_folder = '/content/Test_Data'


train_set = train_datagen.flow_from_directory(train_folder,target_size=(50, 50),batch_size=32,class_mode='categorical',color_mode='rgb',seed=5)

test_set = test_datagen.flow_from_directory(test_folder,target_size=(50, 50),batch_size=32,class_mode='categorical',color_mode='rgb',seed=5)
```
