```
import os
import cv2

path = '/content/SIGNS'

files = []
for r, d, f in os.walk(path):
    for file in f:
        if '.jpg' in file:
            files.append(os.path.join(r, file))

for f in files:
    img=cv2.imread(f)
    img=cv2.resize(img,(300,300))
    cv2.imwrite(f, img)
```

```
from keras.layers import Dense, Conv2D, Flatten

from keras.models import Model,Sequential

model=Sequential()
#model.add(Conv2D(12000, kernel_size=3,strides=2, input_shape=(300,300,3), activation='relu'))
model.add(Conv2D(80,kernel_size=3, strides=2, input_shape=(300,300,3), activation='relu'))
model.add(Conv2D(40,kernel_size=3, strides=2, activation='relu'))
model.add(Conv2D(10,kernel_size=3,activation='relu'))
#model.add(Conv2D(50, kernel_size=3,activation='relu'))
#model.add(Conv2D(10,kernel_size=3, activation='relu'))
#model.add(Conv2D(50,kernel_size=3, activation='relu'))
model.add(Flatten())
model.add(Dense(1024,activation='relu'))
model.add(Dense(512,activation='relu'))
model.add(Dense(256,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(12,activation='relu'))
model.add(Dense(6,activation='softmax'))
model.summary()
```

```
from keras.optimizers import Adam

adam = Adam(epsilon = 0.01)
model.compile(optimizer=adam, loss='categorical_crossentropy', metrics = ['accuracy'])
```

```
from keras.preprocessing.image import ImageDataGenerator

test_datagen = ImageDataGenerator(rescale=1./255)
train_datagen = ImageDataGenerator(rescale=1./255)

#print(test_datagen)

train_folder = '/content/SIGNS/train_signs' 
test_folder = '/content/SIGNS/test_signs'


train_set = train_datagen.flow_from_directory(train_folder,
                                              classes=['0','1','2','3','4','5'],
                                              target_size=(300, 300),
                                              batch_size=32,class_mode='categorical',
                                              color_mode='rgb',
                                              seed=5)

test_set = test_datagen.flow_from_directory(test_folder,
                                            classes=['0','1','2','3','4','5'],
                                            target_size=(300, 300),
                                            batch_size=32,
                                            class_mode='categorical',
                                            color_mode='rgb',
                                            seed=5)
```

```
history = model.fit_generator(train_set,steps_per_epoch=100,epochs=15,validation_data=test_set,validation_steps=30)
```
