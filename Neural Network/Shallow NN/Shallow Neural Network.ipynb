```
import pandas as pd
import numpy as np
```

```
Dataset=pd.read_csv('/content/train.csv')
```

```
Dataset.head()
#print(train.price_range.unique())
#print(train.shape)
```

```
#print(Dataset.columns)
Dataset_y=Dataset[['price_range']]
print(len(Dataset_y))
l=np.zeros((2000,4))
row=0
for i in Dataset_y['price_range']:
  l[row,i]=1
  row+=1
print(l)
Dataset_y=pd.DataFrame(l)
```

```
Dataset_x=Dataset.drop(columns=['price_range'])
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(Dataset_x,Dataset_y,test_size = 0.1)
```

```
from keras.models import Sequential
from keras.layers import Dense
model=Sequential()
model.add(Dense(16, input_dim=20, activation='relu'))
model.add(Dense(12, activation='relu'))
model.add(Dense(4, activation='softmax'))
model.summary()
```

```
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```
model.fit(x_train,y_train,batch_size=10, epochs=50)
```

```
y_pred=model.predict(x_test,batch_size=10)
l=np.zeros((len(y_test),4))
row=0
for i in range(0,len(y_pred)):
    #print(y_pred[i])
    maxval=0
    maxindex=0
    for j in y_pred[i]:
      c=0
      if j>maxval:
        maxval=j
        maxindex=c
    l[row,maxindex]=1
    row+=1
print(l)
```

```
def maxelement(arr): 
      
    # get number of rows and columns 
    no_of_rows = len(arr) 
    no_of_column = len(arr[0]) 
      
    for i in range(no_of_rows): 
          
        # Initialize max1 to 0 at beginning 
        # of finding max element of each row 
        max1 = 0
        for j in range(no_of_column): 
            if arr[i][j] > max1 : 
                max1 = arr[i][j] 
                  
        # print maximum element of each row 
        print(max1)
```

```

_, accuracy = model.evaluate(x_test, y_test)
print('Accuracy: %.2f' % (accuracy*100))
```
